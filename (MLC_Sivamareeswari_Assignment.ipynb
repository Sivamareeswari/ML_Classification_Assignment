```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
#Changing categorial columns to Binary format by using one-hot encoding
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
#Split the input and output
indep=dataset.drop("classification_yes", axis=1)
dep=dataset['classification_yes']
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
```

```python
#======================================
#SVM Model
#======================================
from sklearn.svm import SVC

# Define parameter grid
param_grid_SVC = {
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['auto', 'scale'],
    'C': [10, 100, 1000, 2000, 3000]
}

#  Define SVC with probability=True
svc_model = SVC(probability=True)

# Set up GridSearchCV
grid_SVC = GridSearchCV(
    estimator=svc_model,
    param_grid=param_grid_SVC,
    refit=True,
    verbose=3,
    n_jobs=-1,
    scoring='f1_weighted'
)

#  Fit the grid to training data
grid_SVC.fit(X_train, y_train)

#  Predict using the best model
grid_predictions_SVC = grid_SVC.predict(X_test)
```

```python

cm_SVC = confusion_matrix(y_test, grid_predictions_SVC)
print("The confusion Matrix:\n",cm_SVC)
clf_report_SVC = classification_report(y_test, grid_predictions_SVC)
re=grid_SVC.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
# 6. Evaluate the model
f1_macro_SVC=f1_score(y_test,grid_predictions_SVC,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid_SVC.best_params_),f1_macro_SVC)

print("Accuracy for SVC Model:", accuracy_score(y_test, grid_predictions_SVC))
print("Classification Report for SVC Model:\n", clf_report_SVC)

# 7. ROC AUC Score (for binary classification only)
y_proba_SVC = grid_SVC.best_estimator_.predict_proba(X_test)[:, 1]
auc_SVC = roc_auc_score(y_test, y_proba_SVC)
print("ROC AUC Score for SVC:", auc_SVC)
```

```python
#======================================
#Logistic Regression Model
#======================================
from sklearn.linear_model import LogisticRegression

param_grid_Logistic = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'saga'],
             'penalty':['l2']} 

grid_Logistic = GridSearchCV(LogisticRegression(), param_grid_Logistic, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid_Logistic.fit(X_train, y_train) 
```

```python
#  Predict using the best model
grid_predictions_Logistic = grid_Logistic.predict(X_test)
```

```python
cm_logistic = confusion_matrix(y_test, grid_predictions_Logistic)
print("The confusion Matrix:\n",cm_logistic)
# print classification report 
from sklearn.metrics import classification_report
clf_report_logistic = classification_report(y_test, grid_predictions_Logistic)
```

```python
f1_macro_Logistic=f1_score(y_test,grid_predictions_Logistic,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid_Logistic.best_params_),f1_macro_Logistic)

# 6. Evaluate the model
print("Accuracy for Logistic Model:", accuracy_score(y_test, grid_predictions_Logistic))
print("Classification Report for Logistic Model:\n", clf_report_logistic)

# 7. ROC AUC Score (for binary classification only)
y_proba_Logistic = grid_Logistic.best_estimator_.predict_proba(X_test)[:, 1]
auc_Logistic = roc_auc_score(y_test, y_proba_Logistic)
print("ROC AUC Score for Logistic:", auc_Logistic)
```

```python
#======================================
#Random Forest Classification Model
#======================================
from sklearn.ensemble import RandomForestClassifier

param_grid_RF = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]} 



grid_RF = GridSearchCV(RandomForestClassifier(), param_grid_RF, refit = True, verbose = 3,n_jobs=-1,scoring='f1') 
   
# fitting the model for grid search 
grid_RF.fit(X_train, y_train) 
print(grid_RF.best_params_) 
```
